```python
#import required libraries

import tensorflow as tf # importing tensorflow as tf
from tensorflow import keras
import numpy as np

import pandas

import matplotlib.pyplot as plt


print("hello")
```

```python
#loading the fashion Mnist dataset from keras adatset
fashion_data=keras.datasets.fashion_mnist

((train_data,train_labels),(test_data,test_labels))= fashion_data.load_data()
print("hello")
```

```python
#define target variables classes
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
#analyzing the data

print(train_data.shape)
print(test_data.shape)
```

```python
# displaying 20 images of train data with labels
plt.figure(figsize=(10,10))
for i in range(0,20):
    plt.subplot(5,5,i+1)
    plt.imshow(train_data[i])
    plt.title(class_names[(train_labels[i])])
    plt.xticks([])
    plt.yticks([])
```

```python
train_data=train_data/255.0
test_data=test_data/255.0
```

```python
def con_mod(features,labels,mode):
    input_layer=tf.reshape(features["x"],[-1,28,28,1])
    
    conv = tf.layers.conv2d(input=input_layer,
                          filters=32,
                          kernel_size=[5,5],
                          padding='same',
                          activation=tf.nn.relu
                          )
    pool=tf.layers.max_pooling2d(inputs=conv,poolsize=[2,2],strides=2)
    flatten=tf.reshape(pool,[-1,3*3*128])
    dense=tf.layers.dense(input=flatten,units=1024,activation=tf.nn.relu)
    olayer=tf.layers.dense(input=dense,units=1)
    
    predictions={
        "classes":tf.argmax(input=olayer,axis=1),
        "probabilities":tf.nn.softmax(input=olayer,name='softmax_tensor')
    }
    
    print("hello")
```

```python
classifier= tf.estimator.Estimator(model_fn=con_mod)
```

```python
%Train the model

train_input_fn=tf.estimator.inputs.numpy_input_fn(
    x={"x":train_data},
    y=train_labels,
    batch_size=100,
    num_epochs=None,
    shuffle=True)
classifier.train(input_fn=train_input_fn,steps=1500)
```
